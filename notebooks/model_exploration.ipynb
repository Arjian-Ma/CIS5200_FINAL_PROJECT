```python
# Load the data splits
print("=== Loading Data Splits ===")
try:
    train_data, val_data, test_data = load_data_splits("data/splits")
    print(f"✅ Data splits loaded successfully!")
    print(f"Train: {len(train_data)} rows, {train_data['match_id'].nunique()} matches")
    print(f"Val: {len(val_data)} rows, {val_data['match_id'].nunique()} matches")
    print(f"Test: {len(test_data)} rows, {test_data['match_id'].nunique()} matches")
    
    # Check data shape and columns
    print(f"\nData shape: {train_data.shape}")
    print(f"Columns: {len(train_data.columns)}")
    print(f"Sample columns: {list(train_data.columns[:10])}")
    
except FileNotFoundError as e:
    print(f"❌ Error: {e}")
    print("Please run the dataloader.py script first to create data splits")
```

```python
# Create LSTM datasets
print("=== Creating LSTM Datasets ===")

# Define parameters
sequence_length = 15
batch_size = 32
target_cols = ['Total_Gold_Difference']

# Create datasets
print("Creating training dataset...")
train_dataset = LoLSequenceDataset(
    data=train_data,
    sequence_length=sequence_length,
    target_cols=target_cols,
    fit_scaler=True,
    remove_leakage=True
)

print("Creating validation dataset...")
val_dataset = LoLSequenceDataset(
    data=val_data,
    sequence_length=sequence_length,
    target_cols=target_cols,
    scaler=train_dataset.scaler,  # Use same scaler
    fit_scaler=False,
    remove_leakage=True
)

print("Creating test dataset...")
test_dataset = LoLSequenceDataset(
    data=test_data,
    sequence_length=sequence_length,
    target_cols=target_cols,
    scaler=train_dataset.scaler,  # Use same scaler
    fit_scaler=False,
    remove_leakage=True
)

print(f"✅ Datasets created successfully!")
print(f"Train dataset: {len(train_dataset)} sequences")
print(f"Val dataset: {len(val_dataset)} sequences")
print(f"Test dataset: {len(test_dataset)} sequences")
```

```python
# Check dataset shapes and structure
print("=== Dataset Shape Analysis ===")

# Get a sample from training dataset
sample = train_dataset[0]
print(f"Sample data structure:")
print(f"  - Sequence shape: {sample['sequence'].shape}")
print(f"  - Target shape: {sample['target'].shape}")
print(f"  - Length: {sample['length']}")
print(f"  - Match ID: {sample['match_id']}")

# Check feature dimensions
print(f"\nFeature analysis:")
print(f"  - Number of features: {sample['sequence'].shape[1]}")
print(f"  - Sequence length: {sample['sequence'].shape[0]}")
print(f"  - Target dimensions: {sample['target'].shape[0]}")

# Check a few more samples
print(f"\nMultiple samples:")
for i in range(min(3, len(train_dataset))):
    sample = train_dataset[i]
    print(f"  Sample {i}: seq={sample['sequence'].shape}, target={sample['target'].shape}, length={sample['length']}")
```

```python
# Create DataLoaders with custom collate function
print("=== Creating DataLoaders ===")

# Create DataLoaders
train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=0,  # Set to 0 for notebook compatibility
    pin_memory=False,
    drop_last=True,
    collate_fn=collate_fn
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=0,
    pin_memory=False,
    drop_last=False,
    collate_fn=collate_fn
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=0,
    pin_memory=False,
    drop_last=False,
    collate_fn=collate_fn
)

print(f"✅ DataLoaders created successfully!")
print(f"Train loader: {len(train_loader)} batches")
print(f"Val loader: {len(val_loader)} batches")
print(f"Test loader: {len(test_loader)} batches")
```

```python
# Test DataLoader and check batch shapes
print("=== Testing DataLoader Batches ===")

# Get a batch from training loader
for batch_idx, batch in enumerate(train_loader):
    print(f"Batch {batch_idx} structure:")
    print(f"  - Sequences shape: {batch['sequences'].shape}")
    print(f"  - Targets shape: {batch['targets'].shape}")
    print(f"  - Lengths shape: {batch['lengths'].shape}")
    print(f"  - Match IDs: {batch['match_ids'][:5]}...")  # Show first 5 match IDs
    
    # Check if all sequences in batch are from same match
    unique_matches = set(batch['match_ids'])
    print(f"  - Unique matches in batch: {len(unique_matches)}")
    if len(unique_matches) == 1:
        print(f"  - ✅ All sequences from same match: {list(unique_matches)[0]}")
    else:
        print(f"  - ⚠️  Mixed matches in batch: {list(unique_matches)[:3]}...")
    
    # Check sequence lengths
    print(f"  - Sequence lengths: {batch['lengths'].numpy()[:5]}...")
    print(f"  - Min length: {batch['lengths'].min().item()}")
    print(f"  - Max length: {batch['lengths'].max().item()}")
    
    # Only test first batch
    break

print(f"\n✅ DataLoader test completed!")
```

```python
# Test LSTM input preparation
print("=== LSTM Input Preparation Test ===")

# Get a batch and prepare for LSTM
batch = next(iter(train_loader))
sequences = batch['sequences']
lengths = batch['lengths']

print(f"Raw batch shapes:")
print(f"  - Sequences: {sequences.shape}")  # (batch_size, max_seq_len, n_features)
print(f"  - Lengths: {lengths.shape}")      # (batch_size,)

# Test packing for LSTM efficiency
from torch.nn.utils.rnn import pack_padded_sequence

# Pack sequences for LSTM
packed_sequences = pack_padded_sequence(
    sequences, 
    lengths, 
    batch_first=True, 
    enforce_sorted=False
)

print(f"\nPacked sequences:")
print(f"  - Type: {type(packed_sequences)}")
print(f"  - Data shape: {packed_sequences.data.shape}")
print(f"  - Batch sizes: {packed_sequences.batch_sizes}")

# Test unpacking
from torch.nn.utils.rnn import pad_packed_sequence
unpacked_sequences, unpacked_lengths = pad_packed_sequence(
    packed_sequences, 
    batch_first=True
)

print(f"\nUnpacked sequences:")
print(f"  - Shape: {unpacked_sequences.shape}")
print(f"  - Lengths: {unpacked_lengths}")

print(f"\n✅ LSTM input preparation test completed!")
```
